```python
res =requests.get(url).content
```

```python
soup =BeautifulSoup(res,'html.parser')
```

```python
infos = soup.findAll('div',class_="vendor-info")
```

```python
c=1
```

```python
planners = []
locations = []
reviews = []
ratings = []
prices = []
```

```python
for info in infos:
    planner = info.find('a',class_='vendor-detail')
    #print(c,'-',planner.text.strip())
    planners.append(planner.text.strip())
    
    rating = info.find('span',class_='StarRating')
    #print(c,'-',rating.text.strip())
    ratings.append(rating.text.strip())
    
    location =info.find('p',class_='vendor-detail')
    #print(c,'-',location.text.strip())
    locations.append(location.text.strip())
    
    reviews =info.find('span',class_='review-cnt')
    #print(c,'-',reviews.text.strip())
    reviews.append(reviews.text.strip())
    
    price = info.find('span',class_='')
    #print(price.text.strip())
    prices.append(price.text.strip())
    
    #print('')
    c+=1
```

```python
data ={
    'Planner':planners,
    'Ratings':ratings,
    'Locations':locations,
    'Reviews':reviews,
    'Price':prices}

data
```
