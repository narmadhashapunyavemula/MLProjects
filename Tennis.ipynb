# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Reading dataset

```python
data= pd.read_csv("C:/Users/shapunya vemula/Desktop/dataset/tennis.csv")
```

# Performing EDA

```python
data.head()
```

```python
data['outlook'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data["outlook"] = le.fit_transform(data["outlook"])
data["temp"] = le.fit_transform(data["temp"])
data["windy"] = le.fit_transform(data["windy"])
data["humidity"] = le.fit_transform(data["humidity"])
data["play"] = le.fit_transform(data["play"])
```

```python
data.head()
```

```python
x=data.drop('play',axis=1).values
y=data['play'].values
```

```python
import seaborn as sns
sns.pairplot(data.iloc[:,:-1])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

# performing ID3

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=5)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

# Visualizing the tree

```python
from sklearn import tree
```

```python
fig=plt.figure(figsize=(16,12))
a=tree.plot_tree(model,feature_names=data.columns,class_names=['Not played','Played'],filled=True)
```

# Performing CART

```python
model = DecisionTreeClassifier(criterion='gini',max_depth=5)
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
